Problem Statement: Given an array of integers arr[] and an integer target.

1st variant: Return YES if there exist two numbers such that their sum is equal to the target. Otherwise, return NO.

2nd variant: Return indices of the two numbers such that their sum is equal to the target. Otherwise, we will return {-1, -1}.

```python
def two_sum(arr,target):
    n = len(arr)
    for i in range(n):
        for j in range(n):
            if arr[i]+arr[j] == target:
                index = [i,j]
                return "YES",index
    return "NO",[-1,-1]
            
two_sum([2,6,5,8,11],target = 14)
two_sum([2,6,5,8,11], target = 15)
## 
inedx = two_sum([2,6,5,8,11],target = 14)
print(inedx)
two_sum([2,6,5,8,11], target = 15)
```

Problem Statement: Given an integer array nums, find the subarray 
with the largest sum and return the sum of the elements present in that subarray.
A subarray is a contiguous non-empty sequence of elements within an array

```python

def large_sum(arr):
    n = len(arr)
    max_sum = float('-inf')
    k=n
    for i in range(n):
        curr_sum = 0
       
        for j in range(k):
            curr_sum += arr[j]
        max_sum = max(curr_sum,max_sum)
        k-=1
    return max_sum
print(large_sum([2, 3, 5, -2, 7, -4]))
print(large_sum([-2, -3, -7, -2, -10, -4]))
```

 There’s an array ‘A’ of size ‘N’ with an equal number of positive and negative elements. Without altering the relative order of positive and negative elements, you must return an array of alternately positive and negative values.

```python
def alter_order(arr):
    n = len(arr)
    neg = []
    pos = []
    new = []
    for i in range(n):
        if arr[i]<0:
            neg.append(arr[i])
        else:
            pos.append(arr[i])
    neg.sort()
    pos.sort()
    for i in range(n):
        if i%2==0:
            new.append(pos[int(i/2)])
        else:
            new.append(neg[int(i/2)])
    return new 
print(alter_order([1,2,-4,-5]))
print(alter_order([1,2,-3,-1,-2,3]))
print(alter_order([1,4,-3,-1,-2,3]))
```

Given an array Arr[] of integers, rearrange the numbers of the given array into the lexicographically next greater permutation of numbers.

If such an arrangement is not possible, it must rearrange to the lowest possible order (i.e., sorted in ascending order)

```python
from itertools import permutations
def lex_order(arr):
    perms = list(permutations(arr))
    n= len(arr)
    for i in  range(n):
        if arr == list(perms[-1]):
            return perms[0]
        elif arr == list(perms[i]):
            return perms[i+1]
print(lex_order([1,2,3]))
```

Given an array nums of n integers.

Return the length of the longest sequence of consecutive integers. The integers in this sequence can appear in any order.

```python
def cons_seq(arr):
    if not arr:
        return 0

    arr.sort()

    longest = 1
    count = 1

    for i in range(1, len(arr)):

        # Skip duplicates
        if arr[i] == arr[i-1]:
            continue

        # Consecutive number
        elif arr[i] == arr[i-1] + 1:
            count += 1

        # Break in sequence
        else:
            longest = max(longest, count)
            count = 1

    longest = max(longest, count)

    return longest


print(cons_seq([0, 3, 7, 2, 5, 8, 4, 6, 0, 1]))
print(cons_seq([100, 4, 200, 1, 3, 2]))
print(cons_seq([0, 3, 7, 2, 5, 8, 4, 6, 0, 1]))
print(cons_seq( [1,2,3,10,11]))
```

 ## @
 Given a matrix if an element in the matrix is 0 then you will have to set its entire column and row to 0 and then return the matrix..

```python
def zero_mat(matrix):
    row = len(matrix)
    col = len(matrix[0])

    zero_row = set()
    zero_col = set()

    for i in range(row):
        for j in range(col):
            if matrix[i][j] == 0:
                zero_row.add(i)
                zero_col.add(j)
    
    for i in zero_row:
        matrix[i] = [0]*col

    for c in zero_col:
        for r in range(row):
            matrix[r][c] = 0
    return matrix
print(zero_mat([[1,1,1],[1,0,1],[1,1,1]]))
print(zero_mat([[0,1,2,0],[3,4,5,2],[1,3,1,5]]))
                
```

Given an N * N 2D integer matrix, rotate the matrix by 90 degrees clockwise. The rotation must be done in place, meaning the input 2D matrix must be modified directly..

```python
def right_rot(matrix):
    n = len(matrix)
    new =[[0]*n for _ in range(n)]
    for j in range(n):
        for i in range(n):
            new[j][n-i-1] = matrix[i][j]
    return new
print(right_rot([[0, 1, 1, 2], [2, 0, 3, 1], [4, 5, 0, 5], [5, 6, 7, 0]]))
```

Given an integer array nums of size n. Return all elements which appear more than n/3 times in the array. The output can be returned in any order.

```python
def num_count(arr):
    n = len(arr)
    k = n/3
    op = set()
    for i in range(n):
        count = 0
        for j in range(n):
            if arr[i]==arr[j]:
                count += 1
        if count>=k:
                op.add(arr[i])
    return list(op)
print(num_count([1, 2, 1, 1, 3, 2, 2]))


```

Problem Statement: Given an array of N integers, your task is to find unique triplets that add up to give a sum of zero. In short, you need to return an array of all the unique triplets [arr[a], arr[b], arr[c]] such that i!=j, j!=k, k!=i, and their sum is equal to zero.

```python
def tree_sum(arr):
    n = len(arr)
    s = 0
    op = set() 
    for i in range(n):
        for j in range(i+1,n):
            for k in range(j+1,n):
                sum = arr[i]+arr[j]+arr[k]
                if i!=j and j!=k and i!=k:
                    if sum == s :
                        op.add(tuple(sorted([arr[i],arr[j],arr[k]])))
    return list(op)
print(tree_sum( [-1,0,1,2,-1,-4]))
print(tree_sum( [-1,0,1,0]))

```

 Given an array of N integers, your task is to find unique quads that add up to give a target value. In short, you need to return an array of all the unique quadruplets [arr[a], arr[b], arr[c], arr[d]] such that their sum is equal to a given target.

```python
def four_sum(arr,target):
    n = len(arr)
    sum = 0
    op = set() 
    for i in range(n):
        for j in range(i+1,n):
            for k in range(j+1,n):
                for m in range(k+1,n):
                    sum = arr[i]+arr[j]+arr[k]
                    if sum == target:
                        op.add(tuple(sorted([arr[i],arr[j],arr[k],arr[m]])))
    return op
print(four_sum([1,0,-1,0,-2,2], target = 0))
```

Problem Statement: Given an array containing both positive and negative integers, we have to find the length of the longest subarray with the sum of all elements equal to zero.


```python
def zero_sum(arr):
    n = len(arr)
    vel  =[]
    for i in range(n):
        curr_val = 0
        for j in range(i,n):
            curr_val += arr[j]
            if curr_val == 0:
                vel.append(j-i+1)
    return max(vel)
print(zero_sum([9, -3, 3, -1, 6, -5]))
print(zero_sum([6, -2, 2, -8, 1, 7, 4, -10]))

```

Problem Statement: Given an array of intervals where intervals[i] = [starti, endi], merge all overlapping intervals and return an array of the non-overlapping intervals that cover all the intervals in the input.

```python
def merge_intervals(arr):
    n = len(arr)
    new = [arr[0]]
    for i in range(n):
            if new[-1][1] >= arr[i][0]:
                  new[-1][1] = arr[i][1]
            else:
                  new.append(arr[i])
                  
                  

    return new
print(merge_intervals([[1,3],[2,6],[8,10],[15,18]]))
print(merge_intervals([[1,4],[4,5]]))
                
```

 Given an integer array nums of size n containing values from [1, n] and each value appears exactly once in the array, except for A, which appears twice and B which is missing.
Return the values A and B, as an array of size 2, where A appears in the 0-th index and B in the 1st index.

```python
def dupl(arr):
    n = len(arr)
    dup = None
    mis = None
    
    for i in range(n):
        count = 0
        for j in range(n):
            if arr[i] == arr[j]:
                count += 1
        if count>1:
            dup = arr[i]
            
    for i in range(1,n+1):
        if i not in arr:
            mis = i
    return [dup,mis]
print(dupl([3, 5, 4, 1, 1]))
print(dupl([1, 2, 3, 6, 7, 5, 7]))
```

Given an array of N integers, count the inversion of the array (using merge-sort).

Inversion of an array: for all i & j < size of array, if i < j then you have to find pair (A[i],A[j]) such that A[j] < A[i].

```python
def inversion(arr):
    n = len(arr)
    count = 0
    for i in range(n):
        for j in range(i+1,n):
            if arr[j]<arr[i]:
                count+=1
    return count
print(inversion([1,2,3,4,5]))
print(inversion([5,4,3,2,1]))
print(inversion([5,3,2,1,4]))
```

Given an array of numbers, you need to return the count of reverse pairs. Reverse Pairs are those pairs where i<j and arr[i]>2*arr[j].

```python
def rev_inversion(arr):
    n = len(arr)
    count = 0
    for i in range(n):
        for j in range(i+1,n):
            if arr[i]>2*arr[j]:
                count+=1
    return count
print(rev_inversion([1,3,2,3,1]))
print(rev_inversion([3,2,1,4]))
```

Problem Statement: Given an array that contains both negative and positive integers, find the maximum product subarray

```python
def max_prod(arr):
    n = len(arr)
    max_val = float('-inf')
    prod = 1
    for i in range(n):
        
            prod *= arr[j]
        
        max_val = max(prod,max_val)
    return max_val
print(max_prod([1,2,3,4,5,0]))
print(max_prod([1,2,-3,0,-4,-5]))
```
