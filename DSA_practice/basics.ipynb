```python
##Problem Statement: Given an integer N, return true if it is a palindrome else return false.

##A palindrome is a number that reads the same backward as forward. For example, 121, 1331, and 4554 are palindromes because they remain the same when their digits are reversed.
n = input("enter the Nunber")
def plndrm(n):
    if n == n[::-1]:
        return "palindrome"
    else:
        return "Not a palindrpme"
print(plndrm(n))
```

```python
#Problem Statement: Given two integers N1 and N2, find their greatest common divisor.
n1 = int(input("enter the Nunber"))
n2 = int(input("enter the Nunber"))
def gcd(n1,n2):
    n1_set = set([i for i in range(1,n1) if n1% i == 0 ])
    n2_set = set([i for i in range(1,n2) if n2% i == 0 ])
    inter = n1_set.intersection(n2_set)
    n = max(inter)
    return n

print(gcd(n1,n2))

```

```python
"""An Amrstrong number is a number that is equal to the sum of its own digits each raised to the power of the number of digits.
"""
n = input("enter number")
def armstng(n):
    a = len(n)
    j= 0
    for i in n:
        j = j+int(i)**int(a)
    if j == int(n):
        return "yes"
    else:
        return "no"
print(armstng(n))
```

```python
"""
Problem Statement: Given a number X,  print its factorial.

To obtain the factorial of a number, it has to be multiplied by all the whole numbers preceding it. More precisely X! = X*(X-1)*(X-2) … 1
"""
n = int(input("Enter num"))
def fact(n):
    if n==0 or n==1:
        return 1
    else:
        return n*fact(n-1)
print(fact(n))
```

```python
"""
Given an array arr[] of size n-1 with distinct integers in the range of [1, n]. 
This array represents a permutation of the integers from 1 to n with one element missing. 
Find the missing element in the array.
"""
def miss_num(arr):
    n = len(arr)
    for i in range(1,n+1):
        if i not in arr:
            return f"{i} not in given arr"
        
arr = [8, 2, 4, 5, 3, 7, 1]
    
print(miss_num(arr))

arr = [1, 2, 3, 5]
print(miss_num(arr))
```

```python
"""
Given an array that contains only 1 and 0 return the count of maximum consecutive ones in the array
#Example 1:
Input: prices = {1, 1, 0, 1, 1, 1}
Output: 3
Explanation: There are two consecutive 1’s and three consecutive 1’s in the array out of which maximum is 3.

#Example 2:
Input: prices = {1, 0, 1, 1, 0, 1} 
Output: 2
Explanation: There are two consecutive 1's in the array. 
"""

def num_count(arr):
    n = len(arr)
    count = 0
    max_count = 0
    
    for i in range(n):
        if arr[i] == 1 :
            count +=1
            max_count = max(max_count,count)
        else:
            count = 0
    
    return max_count


arr = [1, 1, 0, 1, 1, 1]
print(num_count(arr))
arr= [1, 0, 1, 1, 0, 1]
print(num_count(arr))
```

```python
"""
Given an array nums of size n and an integer k, 
find the length of the longest sub-array that sums to k. 
If no such sub-array exists, return 0.
"""


def sum_count(arr, k):
    n = len(arr)
    max_len = 0

    for i in range(n):
        curr_sum = 0

        for j in range(i, n):
            curr_sum += arr[j]

            if curr_sum == k:
                length = j - i + 1
                max_len = max(max_len, length)

    return max_len


print(sum_count([10, 5, 2, 7, 1, 9], 15))
```
