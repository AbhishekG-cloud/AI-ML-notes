Problem Statement: Given an input string, containing upper-case and lower-case letters, digits, and spaces( ' ' ). A word is defined as a sequence of non-space characters. The words in s are separated by at least one space. Return a string with the words in reverse order, concatenated by a single space.

```python
def rev_str(s):
        words = []
        word = ""
        for ch in s:
            if ch != " ":
                word += ch
            elif word:
                words.append(word)
                word = ""
        if word:
            words.append(word)
        words.reverse()
        return " ".join(words)
print(rev_str("hi abhi you good ? "))
print(rev_str("my name is python  "))
```

Given a string s, representing a large integer, the task is to return the largest-valued odd integer (as a string) that is a substring of the given string s.
The number returned should not have leading zero's. But the given input string may have leading zero.



```python
def odd_num(strr):
    n = len(strr)
    i=n-1
    end = None
    start = None
    while i>=0:
        if int(strr[i]) % 2 == 1:
            end = i
            break
        i -= 1
    for i in range(n):
        if i < end and strr[i] == 0:
            start = i  
    return strr[start:end+1]  
        
print(odd_num("123456"))
print(odd_num("504"))
print(odd_num("875496"))
    
```

Write a function to find the longest common prefix string amongst an array of strings. If there is no common prefix, return an empty string "".

```python
def lon_com(strr):
    n = len(strr)
    strr.sort()
    start = strr[0]
    end = strr[-1]
    ans = []
    for i in range(min(len(start),len(end))):
        if start[i] == end[i]:
            ans.append(start[i])
    return "".join(ans)
print(lon_com(["flower", "flow", "flight"]))
print(lon_com(["apple", "banana", "grape", "mango"]))
print(lon_com( ["interview", "internet", "internal", "interval"]))  
```

Problem Statement: Given two strings s and goal, return true if and only if s can become goal after some number of shifts on s.
A shift on s consists of moving the leftmost character of s to the rightmost position. For example, if s = "abcde", then it will be "bcdea" after one shift.

```python
def rotateString(s: str, goal: str):
     if len(s) != len(goal):
            return False
     for i in range(len(s)):
           nw  = s[i:]+s[:i]
           if nw == goal:
                return True
     return False
print(rotateString("rotation", "tionrota"))
```

Problem Statement: Given two strings, check if two strings are anagrams of each other or not.

```python
def CheckAnagrams(str1, str2):
    # Case 1: when both of the strings have different lengths
    if len(str1) != len(str2):
        return False  # Strings can't be anagrams if lengths are different

    # Sort both strings
    if sorted(str1) == sorted(str2):
        return True  # Strings are anagrams if sorted versions are equal
    return False  # Strings are not anagrams if sorted versions differ
```

Problem Statement: You are given a string s. Return the array of unique characters, sorted by highest to lowest occurring characters.
If two or more characters have same frequency then arrange them in alphabetic order.

```python
def unique_chars(s):
    freq = {}
    for ch in s:
        if ch in freq:
            freq[ch] += 1
        else:
            freq[ch] = 1
    result = sorted(freq.keys(), key=lambda ch: (-freq[ch], ch))
    return result

        
print(unique_chars("tree"))
```
