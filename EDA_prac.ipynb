```python
df.head(2)
```

```python
df["Date"] = df["Date_of_Journey"].str.split("/").str[0]
df["Month"] = df["Date_of_Journey"].str.split("/").str[1]
df["Year"] = df["Date_of_Journey"].str.split("/").str[1]

```

```python
df["Date"] = df["Date"].astype(int)
df["Month"] = df["Month"].astype(int)
df["Year"] = df["Year"].astype(int)
df.info()
```

```python
df.drop("Date_of_Journey",axis=1,inplace= True)
df.head(2)
```

```python
df["Dep_Hour"] = df["Dep_Time"].str.split(":").str[0]
df["Dep_Min"] = df["Dep_Time"].str.split(":").str[1]
df["Dep_Hour"] = df["Dep_Hour"].astype(int)
df["Dep_Min"] = df["Dep_Min"].astype(int)
df.head(2)
```

```python
df.drop("Dep_Time",axis=1,inplace = True)
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x : x.split(" ")[0])
```

```python
df["Arr_Hour"] = df["Arrival_Time"].str.split(":").str[0]
df["Arr_Min"] = df["Arrival_Time"].str.split(":").str[1]

df.head(2)
```

```python
df["Arrival_Time"] = pd.to_datetime(df["Arrival_Time"],errors="coerce")
df["N_Arr_Hour"] = df["Arrival_Time"].dt.hour

df["N_Arr_Min"] = df["Arrival_Time"].dt.minute
df.head(2)
```

```python
df.drop(["Arr_Hour","Arr_Min"],axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df["Duration"] = pd.to_timedelta(df["Duration"],unit='hours',errors="coerce")
df.head(2)
```

```python
df['Duration_mins'] = df['Duration'].dt.total_seconds()/60
df.head(2)
```

```python
df.drop(["Route"],axis=1,inplace=True)
df.head(2)
```

```python
df['Total_Stops'].unique()
```

```python
df["Stops"] = df["Total_Stops"].map({'non-stop':0, '2 stops':2, '1 stop ': 1, '3 stops':3, np.nan : 1, '4 stops':4})
df.head(2)
```

```python
df.drop([""],axis=1,inplace=True)
df.head(2)
```

```python
df["Additional_Info"].value_counts()
```

```python
df["Airline"].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[["Airline",	"Source","Destination"]]).toarray()
encoded_df = pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
encoded_df
```

```python
df = pd.concat([df,encoded_df],axis=1)
df.head()
```
