```python
def bisection(f, a, b, tol=1e-6, max_iter=100):
    if f(a) * f(b) >= 0:
        print("Invalid interval")
        return None
    
    for _ in range(max_iter):
        c = (a + b) / 2
        if abs(f(c)) < tol:
            return c
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return c
```

```python
## secant
def secant(f, x0, x1, tol=1e-6, max_iter=100):
    for _ in range(max_iter):
        if abs(f(x1) - f(x0)) < 1e-10:
            return None
        x2 = x1 - f(x1)*(x1 - x0)/(f(x1) - f(x0))
        if abs(x2 - x1) < tol:
            return x2
        x0, x1 = x1, x2
    return x2
```

```python
def regula_falsi(f, a, b, tol=1e-6, max_iter=100):
    if f(a)*f(b) >= 0:
        return None
    
    for _ in range(max_iter):
        c = (a*f(b) - b*f(a)) / (f(b) - f(a))
        if abs(f(c)) < tol:
            return c
        elif f(a)*f(c) < 0:
            b = c
        else:
            a = c
    return c
```

```python
def newton(f, df, x0, tol=1e-6, max_iter=100):
    x = x0
    for _ in range(max_iter):
        if abs(df(x)) < 1e-10:
            return None
        x_new = x - f(x)/df(x)
        if abs(x_new - x) < tol:
            return x_new
        x = x_new
    return x
```

```python
#New M
def modified_newton(f, df, d2f, x0, tol=1e-6, max_iter=100):
    x = x0
    for _ in range(max_iter):
        denom = df(x)**2 - f(x)*d2f(x)
        if abs(denom) < 1e-10:
            return None
        x_new = x - (f(x)*df(x))/denom
        if abs(x_new - x) < tol:
            return x_new
        x = x_new
    return x
```

```python
#GJ
def gauss_jacobi(A, b, x0, tol=1e-6, max_iter=100):
    import numpy as np
    x = np.array(x0)
    n = len(b)
    
    for _ in range(max_iter):
        x_new = np.zeros(n)
        for i in range(n):
            s = sum(A[i][j]*x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s)/A[i][i]
        if np.linalg.norm(x_new - x) < tol:
            return x_new
        x = x_new
    return x
```

```python
#GS
def gauss_seidel(A, b, x0, tol=1e-6, max_iter=100):
    import numpy as np
    x = np.array(x0)
    n = len(b)
    
    for _ in range(max_iter):
        x_old = x.copy()
        for i in range(n):
            s = sum(A[i][j]*x[j] for j in range(n) if j != i)
            x[i] = (b[i] - s)/A[i][i]
        if np.linalg.norm(x - x_old) < tol:
            return x
    return x
```

```python
def power_method(A, x0, tol=1e-6, max_iter=100):
    import numpy as np
    x = np.array(x0, dtype=float)
    
    for _ in range(max_iter):
        x_new = A @ x
        x_new = x_new / np.linalg.norm(x_new)
        if np.linalg.norm(x_new - x) < tol:
            break
        x = x_new
    
    eigenvalue = (x.T @ A @ x) / (x.T @ x)
    return eigenvalue, x
```

```python
def lagrange(x_vals, y_vals, x):
    n = len(x_vals)
    result = 0
    for i in range(n):
        term = y_vals[i]
        for j in range(n):
            if i != j:
                term *= (x - x_vals[j])/(x_vals[i] - x_vals[j])
        result += term
    return result
```

```python
def forward_diff(f, x, h=1e-5):
    return (f(x+h) - f(x)) / h
```

```python
def backward_diff(f, x, h=1e-5):
    return (f(x) - f(x-h)) / h
```

```python

def central_diff(f, x, h=1e-5):
    return (f(x+h) - f(x-h)) / (2*h)
```

```python
import math

def forward_diff_table(y):
    n = len(y)
    table = [y.copy()]
    
    for i in range(1, n):
        col = []
        for j in range(n - i):
            col.append(table[i-1][j+1] - table[i-1][j])
        table.append(col)
    return table

def newton_forward(x_vals, y_vals, x):
    h = x_vals[1] - x_vals[0]
    u = (x - x_vals[0]) / h
    
    table = forward_diff_table(y_vals)
    
    result = y_vals[0]
    u_term = 1
    
    for i in range(1, len(y_vals)):
        u_term *= (u - (i-1))
        result += (u_term * table[i][0]) / math.factorial(i)
    
    return result
x_vals=[1,2,3,4]
y_vals=[1,8,27,64]
print("Forward:", newton_forward(x_vals, y_vals, 2.5))
```

```python
def backward_diff_table(y):
    n = len(y)
    table = [y.copy()]
    
    for i in range(1, n):
        col = []
        for j in range(n - i):
            col.append(table[i-1][j+1] - table[i-1][j])
        table.append(col)
    
    return table

def newton_backward(x_vals, y_vals, x):
    h = x_vals[1] - x_vals[0]
    u = (x - x_vals[-1]) / h
    
    table = backward_diff_table(y_vals)
    
    result = y_vals[-1]
    u_term = 1
    
    for i in range(1, len(y_vals)):
        u_term *= (u + (i-1))
        result += (u_term * table[i][-1]) / math.factorial(i)
    
    return result

print("Backward:", newton_backward(x_vals, y_vals, 2.5))
```

```python
def divided_diff(x, y):
    n = len(y)
    coef = y.copy()
    
    for j in range(1, n):
        for i in range(n-1, j-1, -1):
            coef[i] = (coef[i] - coef[i-1]) / (x[i] - x[i-j])
    
    return coef

def newton_divided(x_vals, coef, x):
    n = len(coef)
    result = coef[0]
    
    for i in range(1, n):
        term = coef[i]
        for j in range(i):
            term *= (x - x_vals[j])
        result += term
    
    return result

coef = divided_diff(x_vals, y_vals)
print("Divided Difference:", newton_divided(x_vals, coef, 2.5))
```

```python
def gauss_seidel(A, B, X0, tol=1e-6, max_iter=50):
    n = len(B)
    X = np.array(X0, dtype=float) ## initial
    
    print("Iter\t", "  ".join([f"x{i+1}" for i in range(n)]))
    
    for k in range(max_iter):
        X_old = X.copy()
        
        for i in range(n):
            s = sum(A[i][j]*X[j] for j in range(n) if j != i)
            X[i] = (B[i] - s) / A[i][i]
        
        print(f"{k+1}\t", X)
        
        if np.linalg.norm(X - X_old) < tol:
            return X
    
    return X
```

```python
import numpy as np

def gauss_jacobi(A, B, X0, tol=1e-6, max_iter=50):
    n = len(B)
    X = np.array(X0, dtype=float)
    
    print("Iter\t", "  ".join([f"x{i+1}" for i in range(n)]))
    
    for k in range(max_iter):
        X_new = np.zeros(n)
        
        for i in range(n):
            s = sum(A[i][j]*X[j] for j in range(n) if j != i)
            X_new[i] = (B[i] - s) / A[i][i]
        
        print(f"{k+1}\t", X_new)
        
        if np.linalg.norm(X_new - X) < tol:
            return X_new
        
        X = X_new
    
    return X
```
