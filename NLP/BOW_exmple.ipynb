```python
import pandas as pd
messages = pd.read_csv("SMSSpamCollection",sep='\t',names = ['labels','message'])
messages
```

```python
## Data Cleaning ad Preprocessing
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if word not in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
### Create a Bag of words
from sklearn.feature_extraction.text import CountVectorizer
## for binary bag of words enable binary = True
cv = CountVectorizer(max_features=2500,binary=True) ## max_features will take the highest words with highest frequency
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
### Create a Bag of words using n-grams
cv = CountVectorizer(max_features=500,binary=True,ngram_range=(2,2))
X = cv.fit_transform(corpus).toarray()
```
