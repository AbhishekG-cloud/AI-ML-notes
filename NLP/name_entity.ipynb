### 📌 What is Named Entity Recognition (NER)?

**Named Entity Recognition (NER)** is a natural language processing (NLP) technique that automatically identifies and classifies **named entities** in text into predefined categories such as:

* **PERSON** → Names of people (e.g., "Elon Musk")
* **ORGANIZATION** → Company, institution names (e.g., "NASA")
* **LOCATION** → Geographic places (e.g., "India")
* **DATE** → Time references (e.g., "12 May 2025")
* **MONEY**, **GPE** (Geo-political entity), **FACILITY**, etc.

---

### 🧠 Why is NER useful?

It helps computers **understand the meaning** of text more deeply by recognizing real-world objects. Some uses:

* **Information Extraction** (e.g., get company names from news)
* **Question Answering** (e.g., “Who is the President of the US?”)
* **Search Optimization** (e.g., highlighting entities in documents)
* **Text Summarization**, **Chatbots**, **Resume parsing**, etc.

---

### ⚙️ How NER works?

NER is typically done in these steps:

1. **Tokenization** – Split text into words.
2. **Part-of-Speech (POS) Tagging** – Assign grammatical roles (noun, verb, etc.).
3. **Chunking** – Group words into entities using statistical models or rules.
4. **Classification** – Assign entity types (like PERSON, LOCATION, etc.).

---

### 🔍 Example:

```text
"Steve Jobs founded Apple Inc. in California in 1976."
```

NER output:

| Entity     | Type         |
| ---------- | ------------ |
| Steve Jobs | PERSON       |
| Apple Inc. | ORGANIZATION |
| California | LOCATION     |
| 1976       | DATE         |

---

### 🛠️ In NLTK:

NLTK uses a trained model and grammar rules to do this:

```python
from nltk import word_tokenize, pos_tag, ne_chunk

text = "Steve Jobs founded Apple in California."
tokens = word_tokenize(text)
tagged = pos_tag(tokens)
tree = ne_chunk(tagged)

print(tree)
```

This gives a tree structure where named entities are grouped and labeled.
