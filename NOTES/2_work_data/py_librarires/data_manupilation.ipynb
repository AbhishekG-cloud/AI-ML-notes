## Data Visualisation using Seaborn

```python
!pip install seaborn
```

```python
import seaborn as sns
```

```python
## basic plotting
tps_data = sns.load_dataset('tips')
tps_data
```

```python
## creating a scatter plot 
import matplotlib.pyplot as plt
sns.scatterplot(x= "total_bill",y ="tip",data = tps_data)
plt.title("scater plot") ## we are using plt.title("scatter plot") from Matplotlib (plt) while creating 
                            ## a scatter plot using Seaborn (sns) because Seaborn builds on top of Matplotlib.
plt.show()    
```

```python
##line plot
sns.lineplot(x = "size",y ="total_bill",data=tps_data)
plt.title("line plot")
```

```python
## categorical plots
##bar plot
sns.barplot(x="day",y= "total_bill",data=tps_data)
```

```python
## box plot
sns.boxplot(x="day",y= "total_bill",data=tps_data)
```

```python
## violon plot
sns.violinplot(x="day",y= "total_bill",data=tps_data)
```

```python
## pairplot
sns.pairplot(tps_data)
```

```python
##heat map
corr = tps_data[['total_bill',"tip","size"]].corr()
corr
```

```python
sns.heatmap(corr,annot = True,cmap="coolwarm")
```
