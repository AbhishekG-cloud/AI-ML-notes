```python
import sqlite3
```

```python
## connect to a SQL data base
connection = sqlite3.connect("my_data_base")
connection.execute
```

```python
## create a cursor object
cursor = connection.cursor()
```

```python
# create table 
cursor.execute('''
    Create Table If Not Exists employee(
        id Integer primary Key,
        name Text Not NUll,
        age Intger,
        department Text
                )
''')
#Commit the changes
connection.commit()
```

```python
cursor.execute("Select * From employee")
```

```python
cursor.execute(
    """
Insert Into employee(name,age,department)
                values("abhi",22,"data scientist")
"""
)
cursor.execute(
    """
Insert Into employee(name,age,department)
                values("raj",24,"HR")
"""
)
cursor.execute(
    """
Insert Into employee(name,age,department)
                values("rahul",38," senior data scientist")
"""
)
connection.commit()
```
