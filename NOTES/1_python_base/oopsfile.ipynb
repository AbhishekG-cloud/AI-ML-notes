```python
class BankAccount():
    def __init__(self,account,balance=0):
        self.account = account
        self.balance = balance
    def withdraw(self,n):
        new_bal = self.balance - n
        self.balance = new_bal
        return (f"amoount wihdran is {n} and new bal is {new_bal}")
        
    def credit(self,n):
        new_bal = self.balance + n
        self.balance = new_bal
        return (f"amoount added is {n} and new bal is {new_bal}")
    def currentbal(self):
        return (f"current bal is {self.balance}")
acct_1 = BankAccount("abhi",12000)
acct_2 = BankAccount("yash",14000)
print(acct_1.balance)
print(acct_1.withdraw(1000))
print(acct_1.credit(5000))
acct_1.currentbal()
```

```python
##Inheritance
class  Car():
    def __init__(self,window,doors,engintype):
        self.window = window
        self.doors = doors
        self.engintype = engintype
        
    def drive(self):
        print(f"the driver drives the {self.engintype} car")
class Tesla(Car): ##here tesla is child clas whee car is parent class
    def __init__(self, window, doors, engintype,aotumatic):
        super().__init__(window, doors, engintype)        #we supper to call atrribute from parent class
        self.aotumatic = aotumatic
    def aotu_drive(self):
        print(f"car is aotumatic : {self.aotumatic}")
tes1 = Tesla(4,6,"electric",True)
tes1.drive()
tes1.aotu_drive()
```

```python
##MULTIPLE CALSS INHERITANCE 
## WHEN ACLASS INHERITS FROM MORE THAN ONE CLASS
# BASE CLASS 1
class Animal:
    def __init__(self,name):
        self.name = name
    def speakd(self):
        print("asdfghjkl")
# BASE CLASS 2
class Pet:
    def __init__(self,owner):
        self.owner = owner
# DERIVED CLASS
class Dog(Animal,Pet):
    def __init__(self, name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)
    def spek(self):
        print(f"{self.name} speaks woof") 
dg = Dog("bob","raj")
dg.spek()
dg.speakd()
print(f"owner of {dg.name} is {dg.owner}")
```

```python
## CONCEPT OF POLYMORPHISM
#BASE CLASS1
class Animal:
    def speak(self):
        return "sound"
#Derived Class1
class Dog(Animal):
    def speak(self):
        return "woof"
#Derived Class1
class Cat(Animal):
    def speak(self):
        return "meow"
# function to demostrate polymorphism
def animl_speak(animal):
    return animal.speak()
dg = Dog()
Ct = Cat()
print(dg.speak())
print(Ct.speak())
print(animl_speak(dg))
```

```python
# poly with functions and method
#base class

class Shape:
    def area(self):
        return "the area of fig "
#derived class 1 
class REctangle(Shape):
    def __init__(self,width,lenght):
        self.width = width
        self.lenght = lenght
    def area(self):
        return self.width*self.lenght
#derived class 2
class Circle(Shape):
    def __init__(self,radius):
        self.radius = radius
    def area(self):
        return 3.14*self.radius*self.radius
def print_area(shape):
        print(f" area is {shape.area()}") 

rect = REctangle(4,5)
cir = Circle(3)
print_area(rect)
print_area(cir)
```

PROTECTED VARIABLES CAN BE ACCESSED BY DERIVED CLASS ONLY 

```python

class Person:
    def __init__(self,name,age,gender):
        self._name = name         #here they are protected variables          
        self._age = age           #here they are protected variables
        self.gender = gender 
class Employee(Person):
    def __init__(self,name, age,gender):
        super().__init__(name, age, gender)
        
emply = Employee("abhi",22,"male")
print(emply._name)
#dir(person) 
```

```python
class Vector:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def __add__(self,other):
        return Vector(self.x + other.x, self.y + other.y)
    def __sub__(self,other):
        return Vector(self.x - other.x, self.y - other.y)
    def __mul__(self,other):
        return Vector(self.x*other.x,self.y*other.y)
    def __eq__(self,other):
        return Vector(self.x==other.x and self.y==other.y)
    def __str__(self):                            ## when using print (obj) or str(obj)
        return f"vector ({self.x},{self.y})"
    def __repr__(self):                           ## when using direct call in interactive shell
        return f"vector({self.x},{self.y})"
v1 = Vector(2,3)
v2 = Vector(3,4)
v3 = Vector(-5,-7)
print(v1+v2)
v1+v2+v3
```
