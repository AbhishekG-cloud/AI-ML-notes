```python
def mian_wel(msg):
    def  sub_wel():
        print("welcome to world")
        print(msg)
        print("please be kind")
    return sub_wel()
mian_wel("message for you ")
```

```python
def mian_wel(func):
    def  sub_wel():
        print("welcome to world")
        func("advice for you ")
        print("please be kind")
    return sub_wel()
mian_wel(print)
```

```python
def mian_wel(func,lst):
    def  sub_wel():
        print("welcome to world")
        print(func(lst))
        print("please be kind")
    return sub_wel()
lst = [1,2,3,4,5,6,7]
```

```python
mian_wel(len,lst)
```

```python
##decorators
def mian_wel(func):
    def  sub_wel():
        print("welcome to world")
        func()
        print("please be kind")
    return sub_wel
def hello_fun():
    print("hello world")
```

```python
###now to do above operatiopn without calling the mian function we use decorators
@mian_wel
def hello_fun():
    print("hello world")
## here def hello_fun goes as a parameter in mian_wel() and getting executed
```

```python
hello_fun()
```

```python
def my_dec(fun):   # Decorator function
    def wrapper():  # Inner function that wraps the original function
        print("aSDFGH")  # Extra functionality BEFORE the function call
        fun()  # Call the original function
        print("qwertyui")  # Extra functionality AFTER the function call
    return wrapper  # Returns the wrapper function

@my_dec  # This is the same as saying: say = my_dec(say)
def say():
    print("h")

say()  # Calls the modified 'say' function (which is actually 'wrapper')
```
