```python
# Assignment 15: List Intersection

### Write a function that takes two lists and returns a new list containing only the elements that are present in both lists. 
# Print the intersected list.
def list_intersection(lst1, lst2):
    return [x for x in lst1 if x in lst2]

list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]
intersection = list_intersection(list1, list2)
print(f"List 1: {list1}")
print(f"List 2: {list2}")
print(f"Intersection: {intersection}")
```

```python
#CHECKING PASSWORD STRENGHT
def passwrd_check(passwrd):
    if any(charr.isdigit() for charr in passwrd) == False:
        return False
    elif any(charr.isupper() for charr  in passwrd) == False:
        return False
    elif any(charr.islower() for charr  in passwrd) == False:
        return False
    elif any(charr in "!@#$%^&*()_+=-<>?/.,:';[]|" for charr  in passwrd) == False:
        return False
    else:
        return True
print(passwrd_check("Abhishekg@18"))
print(passwrd_check("Abhishek6@"))
print(passwrd_check(str(1111)))
```

```python
# EMAIL CHECK
import re
def emial_check(email):
    pattern = r'^[a-z A-Z 0-9 _.+-]+@[a-z A-Z]+[.]+[a-z A-Z 0-9]+$'
    return re.match(pattern,email) 
print(emial_check("test@exalp.com"))
```
