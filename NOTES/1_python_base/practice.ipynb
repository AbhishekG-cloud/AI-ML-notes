```python
def sum_list(lst):
    intgrs= []
    try:
        for i in lst:
            c = int(i)
            intgrs.append(c)
    except ValueError as e :
        intgrs = None
    return intgrs
print(sum_list(['1', '2', 'three', '4']))  # None
print(sum_list(['1', '2', '3', '4']))  # [1, 2, 3, 4]
```

```python
def int_converter(lst):
    try: 
        intgrs = [int(x) for x in lst]
    except ValueError:
        intgrs = None
    finally:
        print("code executed")
    return intgrs

print(int_converter(['1', '2', 'three', '4']))  # None
print(int_converter(['1', '2', '3', '4']))  # [1, 2, 3, 4]
```

```python
class Insufball(Exception):
    pass
class Acoount:
    def __init__(self,name,balance):
        self.name = name
        self.balance  = balance
    def withdraw(self,ammount):
        try:
            if ammount > self.balance:
                raise Insufball("insuf bal")
            else:
                self.balance -= ammount
                print(f"new bal is {self.balance}")
        except Insufball as e :
            print("insufficient balance")
p1 = Acoount("abhi",10000)
p1.balance
p1.withdraw(8000)
p1.balance
```
