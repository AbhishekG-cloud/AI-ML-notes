FILE HANDLING

```python
with open('sampl.txt','r') as file:
    mat = file.read()
    print(mat) 
```

```python
#reading code line by line
with open('sampl.txt','r') as file:
    for line in file:
        print(line.strip())
```

```python
#WRITTING A FILE (OVERWRITING) HERE IT EMPTIES THE WHOLE FILE FOR WRITING
with open('sampl.txt','w') as file:
    file.write("hwllo  1234\n")
    file.write("my name is 34")
with open('sampl.txt','r') as file:
    for l in file:
            print(l.strip())


```

```python
#WRITTING A FILE (WITHOUT-OVERWRITING) 
with open('sampl.txt','a') as file:
    file.write("append new line\n")
with open('sampl.txt','r') as file:
    for l in file:
            print(l.strip())


```

```python
#WRIITING NAMY LINES
lines = ["hello you\n", "secomd\n","trird line\n"]
with open('sampl.txt','a') as file:
    file.writelines(lines)

```

```python
# BINARY FILES
data = b'\x00\x01\x02\x03'
with open('rough.bin','wb') as file:
    file.write(data)
```

```python
## read the content from a source file and write to a destination file
# copying a text file
with open('sampl.txt','r') as file:
    matrl = file.read()
with open('new_copied.txt','w') as newfile:
    newfile.write(matrl)
```

 ## Reading large files using generators

```python
def read_file(file_path):
    with open(file_path,"r") as file:
        for line in file:
            yield line
file_path = "large.txt"
for i in read_file(file_path):
    print(i.strip())
```
