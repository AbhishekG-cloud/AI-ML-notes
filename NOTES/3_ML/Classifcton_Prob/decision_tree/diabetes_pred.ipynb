```python
import pandas as pd
```

```python
from sklearn.datasets import load_diabetes
dataset = load_diabetes()
print(dataset['DESCR'])
```

```python
print(dataset.data)
print(dataset.target)
```

```python
X = pd.DataFrame(dataset.data,columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
y = dataset['target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
X_train.corr()
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train,y_train)
```

## hyperparamter tunning

```python
params = {'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],'splitter':['best','random'],'max_depth':[1,2,3,4,5,10,15,20,25],'max_features':['auto','sqrt','log2']}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(regressor,param_grid=params,cv=5,scoring='neg_mean_squared_error')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
