```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
X,y = load_iris(return_X_y=True) ## to return input and output vals
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
import seaborn as sns
df = sns.load_dataset('tips')
```

```python
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder() 
```

```python
X = df.drop(['time','size'],axis=1)
```

```python
y = df['time'].copy()
```

```python
y = le3.fit_transform(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
X_train['sex'] = le1.fit_transform(X_train['sex'])
X_train['smoker'] = le2.fit_transform(X_train['smoker'])
```

```python
X_test['sex'] = le1.transform(X_test['sex'])
X_test['smoker'] = le2.transform(X_test['smoker'])
```

```python
X_train
```

```python
## Onehot encoding--- ColumnTrnasformer
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop='first'),[4])],remainder='passthrough')
```

```python
X_train=ct.fit_transform(X_train)
```

```python
X_test=ct.transform(X_test)
```

```python
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()
```

```python
bnb.fit(X_train,y_train)
```

```python
y_pred = bnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
