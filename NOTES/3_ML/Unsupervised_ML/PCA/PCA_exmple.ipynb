```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# load dataset
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
```

```python
cancer_dataset.keys()
```

```python
for i in cancer_dataset.keys():
    print(cancer_dataset[i])
```

```python
df = pd.DataFrame(cancer_dataset['data'],columns=cancer_dataset['feature_names'])
df.head()
```

```python
## standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

## PCA Algo

```python
from sklearn.decomposition import PCA
pca  = PCA(n_components=2)   # n_compemnets will tell to how many features we want aftyer pcada
data_pca = pca.fit_transform(scaled_data)
```

```python
data_pca[:0]
```

```python
pca.explained_variance_
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x = data_pca[:,0],y = data_pca[:,0],c=cancer_dataset['target'],cmap='plasma')
```
