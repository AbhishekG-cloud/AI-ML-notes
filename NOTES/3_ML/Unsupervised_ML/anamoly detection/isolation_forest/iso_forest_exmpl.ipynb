```python
import pandas as pd
df = pd.read_csv("healthcare.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.iloc[:,0],df.iloc[:,1])
```

```python
from sklearn.ensemble import IsolationForest
```

```python
clf = IsolationForest(contamination=0.2) ## contamination is threshold value
clf.fit(df)
pred = clf.predict(df)
```

```python
pred
```

```python
import numpy as np
index = np.where( pred < 0)
index
```

```python
X = df.values
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1])
plt.scatter(X[index,0],X[index,1])   ## index is list of anomaly points
```
